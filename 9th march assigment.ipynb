Q1: Probability Mass Function (PMF) and Probability Density Function (PDF):
Probability Mass Function (PMF):
A PMF is used in the context of discrete random variables. It gives the probability that a discrete random variable takes on a specific value.
The PMF is denoted as P(X = x), where X is the random variable and x is a specific value.
Example: Consider rolling a fair six-sided die. The PMF for the outcome of rolling the die is:
P(X = 1) = 1/6
P(X = 2) = 1/6
P(X = 3) = 1/6
P(X = 4) = 1/6
P(X = 5) = 1/6
P(X = 6) = 1/6
Probability Density Function (PDF):
A PDF is used for continuous random variables. It describes the likelihood of a continuous random variable taking on a particular value within a range.
The PDF is represented as f(x), and the probability of the variable falling within an interval [a, b] is given by the integral of the PDF over that interval.
Example: The PDF of a standard normal distribution is the bell-shaped curve represented by the equation f(x) = (1 / √(2π)) * e^(-x^2/2), where x is a real number.

Q2: Cumulative Density Function (CDF):
The Cumulative Density Function (CDF) describes the probability that a random variable takes on a value less than or equal to a given value.
It is denoted as F(x), and for a continuous variable, it is calculated by integrating the PDF from negative infinity to x.
Example: For the standard normal distribution, the CDF is represented as F(x) = ∫(from -∞ to x) (1 / √(2π)) * e^(-t^2/2) dt. This function gives the probability that a standard normal random variable is less than or equal to x.
The CDF is useful because it allows us to calculate probabilities associated with a random variable taking on values within a specified range.

Q5: Bernoulli Distribution:
The Bernoulli distribution models a binary outcome, where an event can have two possible outcomes, usually labeled as 0 and 1.
It is characterized by a single parameter p, which represents the probability of a success (1) on a single trial.
Example: Coin flipping, where success (1) represents getting heads, and failure (0) represents getting tails.
Difference between Bernoulli and Binomial Distribution:
Bernoulli distribution models a single trial with two outcomes, while the binomial distribution models the number of successes in a fixed number of independent Bernoulli trials.
Bernoulli has one parameter (p), whereas the binomial has two parameters (n and p), where n is the number of trials.

Q7: Uniform Distribution:
The uniform distribution is a probability distribution where all outcomes within a given range are equally likely.
It has a constant probability density within the range and is represented as a rectangular shape on a probability density function graph.
Example: Rolling a fair six-sided die, where each outcome (1 to 6) has an equal probability of 1/6.

Q8: Z-Score:
The z-score (also known as standard score) measures how many standard deviations a data point is away from the mean in a normal distribution.
It allows for the comparison of data points from different normal distributions.
The z-score is calculated as: Z = (X - μ) / σ, where X is the data point, μ is the mean, and σ is the standard deviation.
Importance: It helps in standardizing data and is useful in statistical analysis, hypothesis testing, and identifying outliers.

Q9: Central Limit Theorem (CLT):
The Central Limit Theorem states that the distribution of the sample means of a large enough random sample from any population will be approximately normally distributed, regardless of the population's underlying distribution.
It is significant because it enables the use of normal distribution-based statistical methods even when the original data may not be normally distributed.

Q10: Assumptions of the Central Limit Theorem:
Random Sampling: The samples should be randomly selected from the population of interest.
Independence: The samples should be independent of each other.
Sample Size: The sample size should be sufficiently large (typically n ≥ 30) to ensure the approximation to normality.
Population Distribution: The underlying population distribution should have a finite mean (μ) and a finite standard deviation (σ)
