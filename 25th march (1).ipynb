```python
#Q6
# Load the dataset into a pandas DataFrame
flight_data = pd.read_excel(file_path)

# Convert 'departure_date' to datetime format
flight_data['Date_of_Journey'] = pd.to_datetime(flight_data['Date_of_Journey'])

# Extract the month from 'departure_date'
flight_data['month'] = flight_data['Date_of_Journey'].dt.month

# Calculate the average price for each month
monthly_avg_prices = flight_data.groupby('month')['Price'].mean()

# Plotting the results
plt.figure(figsize=(15, 8))
monthly_avg_prices.plot(kind='bar', color='skyblue')
plt.title('Average Flight Prices by Month')
plt.xlabel('Month')
plt.ylabel('Average Price')
plt.xticks(rotation=0)
plt.show()
```

```python
#Q7
# Load the dataset into a pandas DataFrame
flight_data = pd.read_excel(file_path)

# Convert 'departure_date' to datetime format
flight_data['Date_of_Journey'] = pd.to_datetime(flight_data['Date_of_Journey'])

# Extract features for analysis
features_to_analyze = ['Date_of_Journey', 'Airline', 'Destination', 'Price']

# Pairplot to visualize relationships between numerical features
sns.pairplot(flight_data[features_to_analyze])
plt.show()

# Line chart for time-based analysis
plt.figure(figsize=(12, 6))
sns.lineplot(x='Date_of_Journey', y='Price', data=flight_data)
plt.title('Trends in Flight Prices over Time')
plt.xlabel('Departure Date')
plt.ylabel('Price')
plt.show()

# Boxplot for seasonal analysis
plt.figure(figsize=(10, 6))
sns.boxplot(x='Destination', y='Price', data=flight_data)
plt.title('Seasonal Trends in Flight Prices')
plt.xlabel('Destination')
plt.ylabel('Price')
plt.show()
```

```python
#Q8

# Load the dataset into a pandas DataFrame
flight_data = pd.read_excel(file_path)

# Route Analysis - Bar chart
plt.figure(figsize=(50,20))
sns.barplot(x='Route', y='Price', data=flight_data)
plt.title('Average Flight Prices by Route')
plt.xlabel('Route')
plt.ylabel('Average Price')
plt.xticks(rotation=45, ha='right')
plt.show()

# Time-Based Analysis - Line chart
plt.figure(figsize=(50,20))
sns.lineplot(x='Date_of_Journey', y='Price', data=flight_data)
plt.title('Trends in Flight Prices over Time')
plt.xlabel('Departure Date')
plt.ylabel('Price')
plt.show()

# Airline-Specific Analysis - Violin plot
plt.figure(figsize=(12,8))
sns.violinplot(x='Airline', y='Price', data=flight_data)
plt.title('Flight Prices by Airline')
plt.xlabel('Airline')
plt.ylabel('Price')
plt.show()

# Flight Duration and Total_Stops - Scatter plot
plt.figure(figsize=(12,8))
sns.scatterplot(x='Duration', y='Price', data=flight_data, hue='Total_Stops')
plt.title('Flight Prices vs. Duration with Total_Stops')
plt.xlabel('Duration (hours)')
plt.ylabel('Price')
plt.show()
```
