```python
#Q1: Five Plots and Their Uses in Seaborn

#Scatter Plot: Used to visualize the relationship between two numerical variables. It helps identify patterns, trends, and correlations.
#Line Plot: Used to display data points connected by straight lines. It's often used to show how a variable changes over time or across categories.
#Bar Plot: Used to compare values across different categories. It's suitable for displaying categorical data with rectangular bars.
#Box Plot: Used to show the distribution of data and identify outliers. It provides information about the median, quartiles, and potential extreme values.
#Histogram: Used to visualize the distribution of a single numerical variable. It breaks the data into bins and shows how many data points fall into each bin.
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

fmri = sns.load_dataset("fmri")
sns.lineplot(data=fmri, x="timepoint", y="signal", hue="event", style="region")
plt.show()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

titanic = sns.load_dataset("titanic")
plt.figure(figsize=(10, 5))
plt.subplot(1, 2, 1)
sns.boxplot(data=titanic, x='pclass', y='age')
plt.subplot(1, 2, 2)
sns.boxplot(data=titanic, x='pclass', y='fare')
plt.tight_layout()
plt.show()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

diamonds = sns.load_dataset("diamonds")
sns.histplot(data=diamonds, x='price', hue='cut', multiple='stack')
plt.show()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

iris = sns.load_dataset("iris")
sns.pairplot(data=iris, hue='species')
plt.show()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

flights = sns.load_dataset("flights")
pivot_flights = flights.pivot_table(index='month', columns='year', values='passengers')
sns.heatmap(data=pivot_flights, cmap='YlGnBu', annot=True, fmt='d')
plt.show()
```
