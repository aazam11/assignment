Q1: The main difference between a t-test and a z-test lies in the type of data they are suitable for and the underlying assumptions:
T-test: It is used when you have a small sample size (typically less than 30) or when the population standard deviation is unknown.
The t-test is based on the t-distribution, which has heavier tails than the normal (z) distribution.
Example scenario: You want to compare the average test scores of two different classes, and you have samples of 15 students from each class. Since the sample size is small, you would use a t-test.
Z-test: It is used when you have a large sample size (typically greater than 30) and know the population standard deviation.
The z-test is based on the normal distribution.
Example scenario: You are testing the mean height of a population of adults, and you have a sample of 100 people.
Since the sample size is large, and you know the population standard deviation, you would use a z-test.

Q2:One-tailed test: In a one-tailed test, you are interested in a specific direction of an effect (e.g., greater than or less than a certain value). You make a hypothesis about the direction of the effect before conducting the test.
Two-tailed test: In a two-tailed test, you are interested in whether there is an effect in either direction (greater than or less than a certain value). You don't specify the direction of the effect in advance.

Q3:Type 1 error (False Positive): This occurs when you reject a true null hypothesis. For example, in a medical test, if you conclude a person has a disease when they don't, it's a Type 1 error.
Type 2 error (False Negative): This occurs when you fail to reject a false null hypothesis. For example, in a medical test, if you conclude a person doesn't have a disease when they actually do, it's a Type 2 error.

Q5: A confidence interval is a range of values used to estimate a population parameter (e.g., the population mean) 
with a certain level of confidence. To calculate a confidence interval, you typically use the formula:

Confidence Interval=Sample Mean ± Margin of Error

The margin of error depends on the confidence level and the standard error of the sample.
Example: Let's say you have a sample of 100 students, 
and you want to estimate the average score of all students in the school with 95% confidence. 
If the sample mean is 85 and the standard error is 3, 
then the confidence interval would be 85 ± (1.96 * 3), which is approximately (79.12, 90.88).
