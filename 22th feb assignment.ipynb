```python
import requests
from bs4 import BeautifulSoup
import csv

url = "https://www.youtube.com/@PW-Foundation/videos"

# Send a GET request to the URL
response = requests.get(url)

# Parse the HTML content
soup = BeautifulSoup(response.content, "html.parser")

# Q1. Extract the video URLs of the first five videos
video_urls = []
video_elements = soup.find_all("a", class_="yt-simple-endpoint style-scope ytd-grid-video-renderer")
for i in range(5):
    video_url = "https://www.youtube.com" + video_elements[i]["href"]
    video_urls.append(video_url)
print("Video URLs:", video_urls)

# Q2. Extract the URLs of the video thumbnails of the first five videos
thumbnail_urls = []
thumbnail_elements = soup.find_all("img", class_="style-scope yt-img-shadow")
for i in range(5):
    thumbnail_url = thumbnail_elements[i]["src"]
    thumbnail_urls.append(thumbnail_url)
print("Thumbnail URLs:", thumbnail_urls)

# Q3. Extract the titles of the first five videos
titles = []
title_elements = soup.find_all("a", class_="yt-simple-endpoint style-scope ytd-grid-video-renderer")
for i in range(5):
    title = title_elements[i].text
    titles.append(title)
print("Titles:", titles)

# Q4. Extract the number of views of the first five videos
views = []
view_elements = soup.find_all("span", class_="style-scope ytd-grid-video-renderer")
for i in range(5):
    view = view_elements[i].text
    views.append(view)
print("Views:", views)

# Q5. Extract the time of posting of the video for the first five videos
timestamps = []
timestamp_elements = soup.find_all("span", class_="style-scope ytd-grid-video-renderer")
for i in range(5, 10):
    timestamp = timestamp_elements[i].text
    timestamps.append(timestamp)
print("Timestamps:", timestamps)

# Save the data in a CSV file
data = zip(video_urls, thumbnail_urls, titles, views, timestamps)
filename = "youtube_data.csv"
with open(filename, "w", newline="", encoding="utf-8") as csvfile:
    writer = csv.writer(csvfile)
    writer.writerow(["Video URL", "Thumbnail URL", "Title", "Views", "Timestamp"])
    writer.writerows(data)
print("Data saved in", filename)
```
